#### Desafio 01: Navegação até um Ponto de Destino usando o robô pioneer 3DX com Aurora

Nesta atividade, você deverá programar o robô para se deslocar até um ponto de destino em duas etapas: primeiro se orientando corretamente usando apenas velocidade angular, e depois se movendo até o ponto usando apenas velocidade linear.

##### Etapas do Controle:

1. Orientação (fase 1):
O robô deve girar sobre seu próprio eixo até estar alinhado com o ponto de destino.
→ Usar apenas **velocidade angular**.
→ Finalizar a fase quando o erro angular for menor que 1 grau.
    
2. Deslocamento (fase 2):
Após estar orientado, o robô deve seguir em linha reta até o destino.
→ Usar apenas **velocidade linear**.
→ A orientação deve ser mantida constante durante o trajeto.

##### Entrada de dados:
A posição e a orientação inicial, e o ponto de destino estarão ser fornecidos no programa.

##### Saída esperada:
- O plot do robô navegando, deixando o rastro de referência.
- Exibir no terminal os erros de posição e erros angulares.
- O sistema deve mudar de fase automaticamente após a orientação correta.

##### Análise dos resultados:
- Discutir a estratégia de chaveamento entre a orientação e o deslocamento, quando o ponto destino está próximo ou distante do ponto de partida.

##### Observações:
- O ambiente é bidimensional $(x, y)$.
- É permitido o uso de bibliotecas matemáticas para cálculo de ângulo.
- Linguagem sugerida: Python ou MATLAB.
