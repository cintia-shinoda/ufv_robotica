# Atividade Avaliativa - AuRoRA

Considere um sistema cooperativo composto por um drone e um robô terrestre Pioneer 3-DX. O drone é comandado manualmente por um operador por meio de um joystick ou teclado, movendo-se para diferentes posições em um mapa tridimensional. A posição instantânea do drone será utilizada como referência para o robô terrestre, que deverá se deslocar de modo a permanecer abaixo do drone.
Desenvolva uma simulação em que:

•	O drone seja movimentado no ambiente 3D por comandos de joystick ou teclado.
•	A posição atual do drone seja continuamente utilizada como referência para o robô Pioneer.
•	O robô Pioneer se mova no plano x−y, buscando permanecer abaixo do drone.
•	O sistema registre, ao longo do tempo:
- Os sinais de controle enviados ao drone;
- A posição 3D do drone;
- Os sinais de controle enviados ao Pioneer;
- A posição do Pioneer no mapa.
